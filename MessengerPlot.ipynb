```
import json
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
import matplotlib.pyplot as plt
% matplotlib inline
```

```
sw = ['có','mà','đi','to', 'là','thế', 'thì', 'sao', 'gì', 'cái','mới', 'dc', 'làm', 'đấy', 'đâu','cũng','vì', 'cứ','is','ok', 'tại','với','trc','of',  'in', 'ms', 'vừa', 'bt', 'đó','nếu', 'do', 'rồi', 'and','bh', 'đc','nh','cơ','thi', 'nên', 'thấy','ấy','chắc', 'các', 'đã', 'chỉ', 'từ','trong', 'mấy','vào', 'sơn', 'ko', 'bị', 'nhưng','chả', 'vẫn','vs','còn','phải','cx', 'đến', 'ng', 'mình', 'đang','chứ', 'lắm', 'như', 'lại', 'cho', 'bạn', 'này', 'nào', 'bảo', 'nó', 'để','của', 'ra', 'nữa', 'được', 'về', 'cả', 'none', 'the', 'quá', 'ai', 'đây',]

mess = []
with open('/content/drive/My Drive/fb messenger/message_1.json', 'r') as f:
  data = json.load(f)
  for item in (data.get('messages')):
    mess.append(str(item.get('content')).encode('latin1').decode('utf8'))

me = " ".join(mess).lower().split(" ")
m = (word_tokenize(" ".join([x for x in me if x not in sw if x.isalnum() if len(x)>1])))

di_freq = nltk.FreqDist(m)
del di_freq['']
li_freq_sorted = sorted(di_freq.items(), key=lambda x: x[1], reverse=True)
di_freq.plot(30, cumulative=False)
```
